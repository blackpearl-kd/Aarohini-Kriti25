```python
# Fill missing `Season` with mode
df['Production Units'] = df['Production Units'].fillna(df['Production Units'].mode()[0])
```

```python
# Fill missing values in 'Production', 'Yield', and 'Area' with mean of the same crop
for col in ['Production', 'Yield', 'Area']:
    df[col] = df.groupby('Crop')[col].transform(lambda x: x.fillna(x.mean()))
# Fill missing values in 'Season' with mode of the same crop
# Function to fill missing values with the mode of the same crop
def fill_mode(series):
    mode_value = series.mode()
    if not mode_value.empty:
        return series.fillna(mode_value.iloc[0])
    else:
        return series  # Return unchanged if there's no mode

df['Season'] = df.groupby('Crop')['Season'].transform(fill_mode)
```

```python
# Function to replace range values with their mean
def replace_range_with_mean(value):
    if isinstance(value, str) and '-' in value:  # Check if the value is a string and contains '-'
        try:
            # Remove commas and split the range into two numbers
            min_val, max_val = value.replace(',', '').split('-')
            mean_val = (int(min_val) + int(max_val)) // 2  # Compute mean
            return mean_val
        except:
            return np.nan  # Return NaN if parsing fails
    return value  # Keep value unchanged if it's not a range

# Apply the function to the column
df['Average Expected Profit per Hectare (INR)'] = df['Average Expected Profit per Hectare (INR)'].apply(replace_range_with_mean)

# Convert column to integer type (if all values are now numeric)
df['Average Expected Profit per Hectare (INR)'] = pd.to_numeric(df['Average Expected Profit per Hectare (INR)'], errors='coerce')

```

```python
import pandas as pd

def allocate_crops(df, state, district, season, total_investment, total_area):
    # Filter the dataset based on the input parameters
    filtered_data = df[
        (df['State'] == state) &
        (df['District'] == district) &
        (df['Season'] == season)
    ].copy()

    if filtered_data.empty:
        return "No crops available for the given inputs."

    # Compute Profit/Investment ratio
    filtered_data['profit_ratio'] = filtered_data['Average Expected Profit per Hectare (INR)'] / filtered_data['Investment per Hectare (INR)']

    # Segregate crops based on nitrogen impact
    crops_nitrogen_plus = filtered_data[filtered_data['Nitrogen Impact'] == 1].sort_values(by='profit_ratio', ascending=False)
    crops_nitrogen_minus = filtered_data[filtered_data['Nitrogen Impact'] == -1].sort_values(by='profit_ratio', ascending=False)
    crops_nitrogen_neutral = filtered_data[filtered_data['Nitrogen Impact'] == 0].sort_values(by='profit_ratio', ascending=False)

    # Lists of crops to iterate over
    nitrogen_crops = [crops_nitrogen_plus, crops_nitrogen_neutral, crops_nitrogen_minus]

    # Initialize tracking variables
    total_area_used = 0
    remaining_investment = total_investment
    crop_allocation = {}

    # Iterate through each crop list in round-robin manner
    crop_indices = [0, 0, 0]  # Track which crop is picked next in each category
    while remaining_investment > 0 and total_area_used < total_area:
        for i, crop_list in enumerate(nitrogen_crops):
            if crop_indices[i] < len(crop_list):  # Check if there are remaining crops in the list
                crop = crop_list.iloc[crop_indices[i]]
                crop_indices[i] += 1  # Move to next crop in the category

                # Extract crop details
                crop_name = crop['Crop']
                investment_per_hectare = crop['Investment per Hectare (INR)']
                area_per_hectare = 1  # Assuming 1 hectare per unit

                # Maximum area that can be allocated for this crop
                max_area_by_investment = remaining_investment / investment_per_hectare
                max_area_by_land = total_area - total_area_used
                area_allocated = min(max_area_by_investment, max_area_by_land, area_per_hectare)

                if area_allocated > 0:
                    # Store allocation
                    if crop_name in crop_allocation:
                        crop_allocation[crop_name] += area_allocated
                    else:
                        crop_allocation[crop_name] = area_allocated

                    # Deduct investment
                    remaining_investment -= (area_allocated * investment_per_hectare)

                    # Update total area used
                    total_area_used += area_allocated

                # Stop if either investment or area is exhausted
                if remaining_investment <= 0 or total_area_used >= total_area:
                    break

    return crop_allocation

# Example usage:
state_input = "Madhya Pradesh"
district_input = "GUNA"
season_input = "Rabi"
total_investment_input = 150000  # Example value
total_area_input = 50  # Example total area in hectares

# Call the function with `df`
crop_allocation_result = allocate_crops(df, state_input, district_input, season_input, total_investment_input, total_area_input)

# Print the output in the required format
print("The area to be used for cropping for each crop based on optimal nitrogen impact and profit is:")
for crop, area in crop_allocation_result.items():
    print(f"{crop}: {area:.2f} hectares")
```
